```python
import pandas as pd
import numpy as np
import cv2 as cv
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from functions import rescaleFrame , chek_shabih , predict_to_array , predict_to_index


from tensorflow import keras
from tensorflow.keras import layers
```

```python
df = pd.read_csv('data.csv')

def str_to_dotayi():

    x = np.zeros((len(X),len(X[0])//2,2))
    y = np.zeros((len(Y),len(Y[0])//2,2))
    for i in range(len(X)):
        for j in range(len(X[i])//2):
            x[i][j] = X[i][j*2], X[i][j*2+1]
    for i in range(len(Y)):
        for j in range(len(Y[i])//2):
            
            y[i][j] = Y[i][j*2], Y[i][j*2+1]
    y_UL = np.zeros((len(X),len(X[0])//2))
    for i in range(len(x)):
        for j in range(len(x[i])):
            if chek_shabih(x[i][j], y[i][0]):
                y_UL[i][j] = 1
    
    y_UR = np.zeros((len(X),len(X[0])//2))
    for i in range(len(x)):
        for j in range(len(x[i])):
            if chek_shabih(x[i][j], y[i][1]):
                y_UR[i][j] = 1
    
    y_DR = np.zeros((len(X),len(X[0])//2))
    for i in range(len(x)):
        for j in range(len(x[i])):
            if chek_shabih(x[i][j], y[i][2]):
                y_DR[i][j] = 1
    
    y_DL = np.zeros((len(X),len(X[0])//2))
    for i in range(len(x)):
        for j in range(len(x[i])):
            if chek_shabih(x[i][j], y[i][3]):
                y_DL[i][j] = 1


    return x, y , y_UL , y_UR , y_DR , y_DL
    



X_columns = df.columns[:230]
Y_columns = df.columns[230:]

X = df[X_columns].values
Y = df[Y_columns].values

x, y , y_UL , y_UR , y_DR , y_DL = str_to_dotayi()
df
```

```python
X_train, X_test,\
y_UL_train,y_UL_test,\
y_UR_train, y_UR_test,\
y_DR_train, y_DR_test,\
y_DL_train, y_DL_test = train_test_split(x, y_UL,y_UR , y_DR , y_DL ,test_size=0.3)
```

```python
sc = StandardScaler()

#X_train = sc.fit_transform(X_train)

#X_test = sc.transform(X_test)
```

```python
x_train = X_train.copy()
for i in range(len(X_train)):
    x_train[i] = X_train[i]/X_train[i].std()

x_test = X_test.copy()
for i in range(len(X_test)):
    x_test[i] = x_test[i]/X_test[i].std()
```

```python
x_input = keras.Input(
    shape=(115,2) , name = "voroodi")
daste = layers.Dense(40, activation="relu" ,name="4daste")(x_input)
daste = layers.Reshape((4600,))(daste)

UP = layers.Dense(100, activation="relu" ,name="UP")(daste)
DO = layers.Dense(100, activation="relu" ,name="DO")(daste)
LE = layers.Dense(100, activation="relu" ,name="LE")(daste)
RI = layers.Dense(100, activation="relu" ,name="RI")(daste)


ul_n  = layers.concatenate([UP,LE],name="ul_n")
ur_n  = layers.concatenate([UP,RI],name="ur_n")
dr_n  = layers.concatenate([DO,RI],name="dr_n")
dl_n  = layers.concatenate([DO,LE],name="dl_n")

ul_n2  = layers.Dense(3000 , activation="relu",name = "ul_pardazeh")(ul_n)
ur_n2  = layers.Dense(3000 , activation="relu",name = "ur_pardazeh")(ur_n)
dr_n2  = layers.Dense(3000 , activation="relu",name = "dr_pardazeh")(dr_n)
dl_n2  = layers.Dense(3000 , activation="relu",name = "dl_pardazeh")(dl_n)

ul_n3 = layers.Dense(800 , activation="relu",name = "ul_pardazeh2")(ul_n2)
ur_n3 = layers.Dense(800 , activation="relu",name = "ur_pardazeh2")(ur_n2)
dr_n3 = layers.Dense(800 , activation="relu",name = "dr_pardazeh2")(dr_n2)
dl_n3 = layers.Dense(800 , activation="relu",name = "dl_pardazeh2")(dl_n2)

ul_n4 = layers.Dense(300 , activation="relu",name = "ul_pardazeh3")(ul_n3)
ur_n4 = layers.Dense(300 , activation="relu",name = "ur_pardazeh3")(ur_n3)
dr_n4 = layers.Dense(300 , activation="relu",name = "dr_pardazeh3")(dr_n3)
dl_n4 = layers.Dense(300 , activation="relu",name = "dl_pardazeh3")(dl_n3)



ul_o  = layers.Dense(115, activation="softmax" , name ="ul_output")(ul_n4)
ur_o  = layers.Dense(115, activation="softmax" , name ="ur_output")(ur_n4)
dr_o  = layers.Dense(115, activation="softmax" , name ="dr_output")(dr_n4)
dl_o  = layers.Dense(115, activation="softmax" , name ="dl_output")(dl_n4)

```

```python


model = keras.Model(
    inputs=x_input,
    outputs=[ul_o,ur_o,dr_o,dl_o])
    
model.compile(optimizer= 'adam',
                  loss = 'categorical_crossentropy',
                  metrics = ['accuracy'])
model.summary()
```

```python
model.fit(
    {"voroodi": x_train}, 
    {"ul_output":y_UL_train, "ur_output" : y_UR_train, "dr_output":y_DR_train, "dl_output":y_DL_train},
            epochs=10)
```

```python
predict_UL = predict_to_index(model.predict(x_test)[0])
predict_UR = predict_to_index(model.predict(x_test)[1])
predict_DR = predict_to_index(model.predict(x_test)[2])
predict_DL = predict_to_index(model.predict(x_test)[3])

array_UL = predict_to_array(model.predict(x_test)[0])
array_UR = predict_to_array(model.predict(x_test)[1])
array_DR = predict_to_array(model.predict(x_test)[2])
array_DL = predict_to_array(model.predict(x_test)[3])

sum = 0
for i in range(len(predict_UL)):
    if chek_shabih(y_UL_test[i] , array_UL[i]) and chek_shabih(y_UR_test[i] , array_UR[i]) and chek_shabih(y_DR_test[i] , array_DR[i]) and chek_shabih(y_DL_test[i] , array_DL[i]):
        sum += 1
print(sum/len(predict_UL))
```
