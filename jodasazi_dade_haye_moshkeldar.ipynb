```python
import cv2 as cv
import numpy as np
import pandas as pd
from  functions import rescaleFrame 
```

```python
df = pd.read_csv('data.csv')
df
```

```python
i = 0
l = []
while i <len(df):
    img = np.zeros((1400,1400,3), np.uint8)

    #pick row i from fd
    row = df[i:i+1]
    for j in range(115):
        cv.circle(img ,(int(row["x"+str(j)]+700) , int(row["y"+str(j)])+700), 7, (255,255,255), -1)
    
    try:
        cv.circle(img ,(int(row["ULx"]+700) , int(row["ULy"])+700), 9, (25,2,255), -1)
        cv.circle(img ,(int(row["URx"]+700) , int(row["URy"])+700), 9, (5,215,255), -1)
        cv.circle(img ,(int(row["DLx"]+700) , int(row["DLy"])+700), 9, (25,255,5), -1)
        cv.circle(img ,(int(row["DRx"]+700) , int(row["DRy"])+700), 9, (255,25,5), -1)
    except:
        cv.putText(img, "entekhab nashide", (300,300), cv.FONT_HERSHEY_SIMPLEX, 1, (255,255,55), 2)

    cv.putText(img , str(i) , (img.shape[1]//4 , img.shape[0]//2) , cv.FONT_HERSHEY_SIMPLEX , 2 , (0,0,255) , 3)

    if len(l) >=3:
        cv.putText(img , str(l[-3:]) , (img.shape[1]//4 , img.shape[0]*3//4) , cv.FONT_HERSHEY_SIMPLEX , 2 , (0,0,255) , 3)
    elif len(l) ==2:
        cv.putText(img , str(l[-2:]), (img.shape[1]//4 , img.shape[0]*3//4) , cv.FONT_HERSHEY_SIMPLEX , 2 , (0,0,255) , 3)
    elif len(l) ==1:
        cv.putText(img , str(l[-1:]), (img.shape[1]//4 , img.shape[0]*3//4) , cv.FONT_HERSHEY_SIMPLEX , 2 , (0,0,255) , 3)
    else:
        cv.putText(img , "no l", (img.shape[1]//4 , img.shape[0]*3//4) , cv.FONT_HERSHEY_SIMPLEX , 2 , (0,0,255) , 3)
    cv.imshow("img" , rescaleFrame(img, 0.4 ))
    

    if cv.waitKey(0) == ord('x'):
        i+=1
    elif cv.waitKey(0) == ord('l'):
        l.append(i)
    elif cv.waitKey(0) == ord('z'):
        i-=1
    elif cv.waitKey(0) == ord('q'):
        break
cv.destroyAllWindows()
```
